```python
WINDOW = 4

def get_angle(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return np.arctan2(y2-y1, x2-x1)

def get_distance(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return np.sqrt((x2-x1)**2 + (y2-y1)**2)

def get_image_angle(lista):
    angles = []
    for i in range(len(lista)-WINDOW):
        angle = get_angle(lista[i], lista[i+WINDOW])
        if angle < 0: angle += 2*np.pi
        if angle > 2*np.pi: angle -= 2*np.pi
        angles.append(round(angle, 3))
    return angles

def find_more_common_angle(lista):
    return max(set(lista), key=lista.count)


h = get_image_angle(points)

angle_ = find_more_common_angle(h)
print(angle_)
basee_point = points[h.index(angle_)]

# plt.hist(h, bins=100)
# plt.show()
```

```python
def get_image(basee_point, angle_, points_):

    mapa = np.zeros((2*2*ESCALA, 4*ESCALA), dtype=np.uint8)

    p = points_[0]
    ang = get_angle(basee_point, p)
    dist = get_distance(basee_point, p)
    x = round(dist*np.cos(ang - angle_ + np.pi)*ESCALA)
    y = round(dist*np.sin(ang - angle_ + np.pi)*ESCALA)
    center = (-y + 2*ESCALA, 2*ESCALA + x)

    for p in points_[1:]:
        ang = get_angle(basee_point, p)
        dist = get_distance(basee_point, p)
        x = round(dist*np.cos(ang - angle_ + np.pi)*ESCALA)
        y = round(dist*np.sin(ang - angle_ + np.pi)*ESCALA)
        mapa[-y + 2*ESCALA, 2*ESCALA + x] = BLACK

    return mapa, center

fixed_map, center = get_image(basee_point, angle_, points)

plt.imshow(fixed_map, cmap='gray')
# plt.imshow(cv2.dilate(fixed_map, kernel), cmap='gray')
```

```python
def find_edges(image):
    top = -1
    bottom = -1
    right = -1
    left = -1

    for i in range(len(image)):
        if top == -1:
            if np.sum(image[i]) != 0:
                top = i
        if bottom == -1:
            if np.sum(image[-i - 1]) != 0:
                bottom = len(image) - i

    for j in range(len(image[0])):
        if left == -1:
            if np.sum(image[:, j]) != 0:
                left = j
        if right == -1:
            if np.sum(image[:, -j - 1]) != 0:
                right = len(image[i]) - j
    return top, bottom, left, right



T, B, L, R = find_edges(fixed_map)
final_mask = fixed_map[T:B, L:R]
center_ = (center[0] - T, center[1] - L)
print(center_)

plt.imshow(final_mask, cmap='gray')
# plt.imshow(cv2.dilate(final_mask, kernel), cmap='gray')
```

```python
def convolve2(image, mask):
    y_ = image.shape[0] - mask.shape[0] + 1
    x_ = image.shape[1] - mask.shape[1] + 1

    out = np.zeros([y_, x_])

    for n in range(mask.shape[0]):
        for m in range(mask.shape[1]):
            out += image[n:n+y_, m:m+x_] * mask[n, m]
    return out
```

```python
def gaussiankernel(size):
    """
    This method creates a symmetric Gaussian kernel.

    :param size: inf number to specify the size of the kernel.
            Typical numbers 3, 5, 9, 11.
    """

    sigma = size/8.5 #  width of the Gaussian distribution

    x, y = np.mgrid[-size // 2 + 1:size // 2 + 1,
           -size // 2 + 1:size // 2 + 1]

    kernel = np.exp(-((x ** 2 + y ** 2) / (2.0 * sigma ** 2)))
    kernel = kernel / kernel.sum()

    return kernel.astype('double')
```

```python
k = 100
ker = gaussiankernel(k) # se genera el kernel de degradación
Go = cv2.filter2D(dark_map.astype('double'), cv2.CV_64F, ker) # se degrada la imagen

plt.imshow(Go,cmap = "gray")
```
