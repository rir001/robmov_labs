```python
def get_angle(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return np.arctan2(y2-y1, x2-x1)

def get_distance(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return np.sqrt((x2-x1)**2 + (y2-y1)**2)

def get_image_angle(lista):
    angles = []
    rounded_angles = []
    for i in range(len(lista)-4):
        angle = get_angle(lista[i], lista[i+4])
        angles.append(angle)
        rounded_angles.append(round(angle, 3))
    return (angles, rounded_angles)

def get_point_with_angle(lista, angle):
    for i in range(len(lista)-2):
        if angle -0.1 < get_angle(lista[i], lista[i+2]) < angle + 0.1:
            return [lista[i], lista[i+2]]

def find_more_common_angle(lista):
    return max(set(lista), key = lista.count)

def most_frequent(List):
    counter = 0
    num = List[0]

    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency > counter):
            counter = curr_frequency
            num = i
    return num

(h, rounded_h) = get_image_angle(points)

for n in range(len(h)):
    if h[n] < 0:
        h[n] = h[n] + np.pi


plt.hist(h, bins=100)
plt.show()

angle_ = most_frequent(rounded_h) 
angle_1 = find_more_common_angle(rounded_h)

print(angle_1, angle_)
basee_point = get_point_with_angle(points, angle_1)[0]
```

```python
def get_image(basee_point, angle_, points_):

    mapa = np.zeros((2*2*ESCALA, 4*ESCALA), dtype=np.uint8)

    p = points_[0]
    ang = get_angle(basee_point, p)
    dist = get_distance(basee_point, p)
    x = int(dist*np.cos(ang - angle_ + np.pi)*ESCALA)
    y = int(dist*np.sin(ang - angle_ + np.pi)*ESCALA)
    center = (-y + 2*ESCALA, 2*ESCALA + x)

    for p in points_[1:]:
        ang = get_angle(basee_point, p)
        dist = get_distance(basee_point, p)
        x = int(dist*np.cos(ang - angle_ + np.pi)*ESCALA)
        y = int(dist*np.sin(ang - angle_ + np.pi)*ESCALA)
        mapa[-y + 2*ESCALA, 2*ESCALA + x] = BLACK

    return mapa, center

fixed_map, center = get_image(basee_point, angle_1, points)

plt.imshow(cv2.dilate(fixed_map, kernel), cmap='gray')
```

```python
def find_edges(image):
    top = -1
    bottom = -1
    right = -1
    left = -1

    for i in range(len(image)):
        if top == -1:
            if np.sum(image[i]) != 0:
                top = i
        if bottom == -1:
            if np.sum(image[-i - 1]) != 0:
                bottom = len(image) - i - 1

    for j in range(len(image[0])):
        if left == -1:
            if np.sum(image[:, j]) != 0:
                left = j
        if right == -1:
            if np.sum(image[:, -j - 1]) != 0:
                right = len(image[i]) - j - 1

    return top, bottom, left, right



# Recorta la imagen de un tamaño 180x180 desde el pixel izq, arriba
T, B, L, R = find_edges(fixed_map)
fm_final_xd = fixed_map[T:B, L:R]
center_ = (center[0] - T, center[1] - L)
print(center_)
# plt.imshow(fm_final_xd, cmap='gray')
plt.imshow(cv2.dilate(fm_final_xd, kernel), cmap='gray')
```

```python
def convolve2(image, mask):
    y_ = image.shape[0] - mask.shape[0] + 1
    x_ = image.shape[1] - mask.shape[1] + 1

    out = np.zeros([y_, x_])

    for n in range(mask.shape[0]):
        for m in range(mask.shape[1]):
            out += image[n:n+y_, m:m+x_] * mask[n, m]
    return out
```
